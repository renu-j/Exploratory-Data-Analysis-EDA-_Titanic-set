```python
# Titanic Dataset - Exploratory Data Analysis (EDA)
1.
# Importing necessary libraries
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')

2.
# Loading the dataset

titanic_data = pd.read_csv("../data/raw/Titanic_Dataset.csv")

titanic_data.head()
3.
#checking dataset Shape
titanic_data.shape
4.
#checking dataset COLUMNS NAMES
titanic_data.columns

5.
#checking dataset INFO
titanic_data.info()

6.
#checking dataset DESCRIBE
titanic_data.describe()

7.
#Misssing Values Analysis
titanic_data.isnull().sum()

8.
#Visualizing Missing Values

plt.figure(figsize=(10,6))
sns.heatmap(titanic_data.isnull(), cbar=False, cmap='viridis')
plt.title('Missing Values Heatmap')
plt.show()

9.
#Data cleaning - Dropping 'Cabin' column due to high number of missing values
titanic_data.drop('Cabin', axis=1, inplace=True)
#Rechecking missing values after dropping 'Cabin'
titanic_data.isnull().sum()

10.
#handling missing values in 'Age' column by filling with median age
titanic_data['Age'].fillna(titanic_data['Age'].median(), inplace=True)
#Rechecking missing values after filling 'Age'
titanic_data.isnull().sum() 

11.#handling missing values in 'Embarked' column by filling with mode
titanic_data['Embarked'].fillna(titanic_data['Embarked'].mode()[0], inplace=True)
#Rechecking missing values after filling 'Embarked' 
titanic_data.isnull().sum()

12.
#Data Visualization and Pattern Analysis
#Survival Distribution
sns.countplot(x='Survived', data=titanic_data)
plt.title('Survival Distribution')
plt.xlabel('Survived (0 = No, 1 = Yes)')
plt.ylabel('Count')
plt.savefig('../images/survival_distribution.png')
plt.show()

13.
#Survival Distribution by Gender vs Survival
sns.countplot(x="Sex", hue="Survived", data=titanic_data)
plt.title(' Gender vs Survival ')
plt.xlabel('Gender')    
plt.ylabel('Count')
plt.legend(title='Survived', loc='upper right', labels=['No', 'Yes'])
plt.savefig('../images/gender_survival_distribution.png')
plt.show()  

14.
#Passenger Class Vs Survival
sns.countplot(x="Pclass", hue="Survived", data=titanic_data)
plt.title('Passenger Class Vs Survival')
plt.xlabel('Passenger Class')   
plt.ylabel('Count')
plt.legend(title='Survived', loc='upper right', labels=['No', 'Yes'])
plt.savefig('../images/class_survival.png')
plt.show()

15.
#Age Distribution
sns.histplot(titanic_data['Age'], bins=30, kde=True)
plt.title('Age Distribution of Passengers')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.savefig('../images/age_distribution.png')
plt.show()

16.
#Correlation Analysis
plt.figure(figsize=(10,6))

sns.heatmap(
    titanic_data.corr(numeric_only=True),
    annot=True,
    cmap="coolwarm",
    linewidths=0.5
)

plt.title("Correlation Heatmap")
plt.savefig('../images/correlation_heatmap.png')
plt.show()

#observations:
# Female passenger had significantly higher survival rate than male passengers.
# Passengers Fare shows a positive correlation with survival,while higher passenger class (lower  socioeconomic status ) is negatively correlated with survival probability.

17.
#Key Insights and Conclusions
### Key Insights

1.
##Gender Impact on Survival  
#Female passengers had a significantly higher survival rate compared to male passengers, indicating priority evacuation policies during the disaster.

2.
#Passenger Class Influence  
#Passengers traveling in first class showed higher survival probability compared to second and third class passengers, suggesting socioeconomic factors influenced rescue outcomes.

3.
#Age Distribution  
#Most passengers were between 20–40 years old. Younger passengers showed slightly better survival trends.

4.
#Fare Relationship  
#Passengers who paid higher fares tended to have higher survival rates, which correlates with higher passenger class.

5. 
#Missing Data Handling  
#Missing values in Age and Embarked were successfully handled using statistical imputation techniques, and the Cabin column was removed due to excessive missing data.

### Conclusions
#This Exploratory Data Analysis (EDA) examined the Titanic dataset to understand survival patterns among passengers. Data cleaning techniques were applied to handle missing values, followed by statistical analysis and visualization to identify meaningful relationships.

#The analysis revealed that gender, passenger class, and fare were strong indicators of survival probability. Visualization techniques helped uncover important patterns and correlations within the dataset.

#This project demonstrates practical skills in data preprocessing, statistical analysis, and data visualization using Python, Pandas, Matplotlib, and Seaborn.

18.
#Save cleaned dataset
titanic_data.to_csv("../data/processed/titanic_cleaned.csv", index=False)

```
